```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('out.png')
h, w = img.shape[:2]
plt.imshow(img);
```

```python
channel_0 = np.zeros_like(img)
for i in range(h):
    for j in range(w):
        channel_0[i, j] = img[i, j, 0] & 1

plt.figure(figsize=(16, 16))
plt.imshow(channel_0 * 255);
```

```python
channel_1 = np.zeros_like(img)
for i in range(h):
    for j in range(w):
        channel_1[i, j] = img[i, j, 1] & 1

plt.figure(figsize=(16, 16))
plt.imshow(channel_1 * 255);
```

```python
channel_2 = np.zeros_like(img)
for i in range(h):
    for j in range(w):
        channel_2[i, j] = img[i, j, 2] & 1

plt.figure(figsize=(16, 16))
plt.imshow(channel_2 * 255);
```
