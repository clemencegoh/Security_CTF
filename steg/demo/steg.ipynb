```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import cv2

from steg import replace, binarize, hide
```

```python
hidden = cv2.imread('lena.png')
hidden = binarize(hidden)
plt.imshow(hidden, cmap='gray');
```

```python
view = cv2.imread('tux.jpg', 0)
plt.imshow(view, cmap='gray');
```

```python
out = hide(view, hidden)
plt.imshow(out, cmap='gray');
```

```python
res = np.zeros_like(out)
h, w = res.shape[:2]

for i in range(h):
    for j in range(w):
        res[i, j] = out[i, j] & 1
        
plt.imshow(res, cmap='gray');
```
